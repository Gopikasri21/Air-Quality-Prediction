```python
import pandas as pd
import numpy as np
import pickle
import streamlit as st
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
dataset = pd.read_csv("AQI Dataset.csv")
```

```python
dataset = dataset.dropna()
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,dtype=int,drop_first = True)
```

```python
dataset
```

```python
Independent = dataset[['PM2.5','PM10','NO2','SO2','CO','O3','Temperature','Humidity','Wind Speed','Country_USA','City_Chennai','City_Delhi','City_Dubai','City_London','City_Los Angeles','City_Mumbai','City_New York','City_Shanghai','City_Sydney']]
```

```python
Independent
```

```python
Dependent = dataset[['AQI']]
```

```python
Dependent
```

```python
X_train,X_test,y_train,y_test = train_test_split(Independent,Dependent,test_size=0.30,random_state=0)
```

```python
model = RandomForestRegressor()
model.fit(Independent,Dependent)
```

```python
X_train
```

```python
Regressor = RandomForestRegressor(n_estimators=200,criterion='squared_error', max_depth=10, min_samples_split=5, random_state=42)
```

```python
Regressor
```

```python
Regressor.fit(X_train,y_train)
```

```python
y_pred = Regressor.predict(X_test)
```

```python
y_pred
```

```python
r_score = r2_score(y_test,y_pred)
```

```python
r_score
```
